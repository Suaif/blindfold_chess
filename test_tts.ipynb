```python
import subprocess
import os

voices = {
    "bryce": r"bryce-medium\en_US-bryce-medium.onnx",
    "hfc_male": r"hfc_male-medium\en_US-hfc_male-medium.onnx"
}

model = "bryce"

# Paths
piper_dir = r"C:\Users\ismas\projects\blindfold_chess\assets\text_to_speech\piper"
model_path = os.path.join(piper_dir, voices[model])

output_folder = r"C:\Users\ismas\projects\blindfold_chess"
output_wav = os.path.join(output_folder, f"output_{model}.wav")

# Text to synthesize
text = "Hello, this is a test using Piper TTS directly from Python."

# Run piper.exe via subprocess
cmd = [
    os.path.join(piper_dir, "piper.exe"),
    "-m", model_path,
    "-f", output_wav
]

subprocess.run(cmd, input=text.encode("utf-8"), check=True)
print(f"✅ Audio written to {output_wav}")
```
